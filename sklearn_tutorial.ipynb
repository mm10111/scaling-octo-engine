```python
# importing the wine dataset from sklearn
from sklearn.datasets import load_wine
wine_dataset = load_wine()
```

```python
# import the data into pandas data frame
import pandas as pd
# read the data into a pandas dataframe
wine_dataframe = pd.DataFrame(wine_dataset.data, columns=wine_dataset.feature_names)
# add the target label for the dataset
wine_dataframe['target'] = wine_dataset.target
# view the top of the dataset
wine_dataframe.head()
```

```python
# view the information about the dataset
wine_dataframe.info()
```

```python
# get descriptive statistics of the dataset
wine_dataframe.describe()
```

Pre-process the data so it is ready for machine learning

```python
# import the standardscaler from sklearn
from sklearn.preprocessing import StandardScaler
# standard scaler function standardizes features to unit variance by subtracting the mean.
# unit variance is when the variance of the dataset is equal to one.
# links for further reading: https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html
# https://www.geeksforgeeks.org/data-science/what-is-zero-mean-and-unit-variance-normalization/

# split the data into features and the label
X = wine_dataframe[wine_dataset.feature_names].copy()
y = wine_dataframe["target"].copy()
# get the scalar and fit on features
scaler = StandardScaler()
scaler.fit(X)

# transform the features
X_scaled = scaler.transform(X.values)

# view the first instance
print(X_scaled[0])
```

```python
# split the data into training and test set
from sklearn.model_selection import train_test_split
# split the data into train and test
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y, train_size=.7, random_state=25)
```

```python
# check if the data is split correctly
print(f"Train size: {round(len(X_train_scaled) / len(X) * 100)}% \n\
Test Size: {round(len(X_test_scaled) / len(X) * 100)}%")
```

```python
# Data is processed and now is ready for the machine learning model
# importing models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

# instantiate the models
logistic_regression = LogisticRegression()
svm = SVC()
decision_tree = DecisionTreeClassifier()

# training the models
logistic_regression.fit(X_train_scaled, y_train)
svm.fit(X_train_scaled, y_train)
decision_tree.fit(X_train_scaled, y_train)
```

```python
# make predictions with each model
log_reg_predictions = logistic_regression.predict(X_test_scaled)
svm_predictions = svm.predict(X_test_scaled)
decision_tree_predictions = decision_tree.predict(X_test_scaled)
```

```python
# models evaluation
from sklearn.metrics import classification_report

# storing model predictions in a dictionary
model_predictions = {
    "Logisitic Regression": log_reg_predictions,
    "Support Vector Machine": svm_predictions,
    "Decision Tree": decision_tree_predictions
}
for model, predictions in model_predictions.items():
  print(f"{model} Results:\n{classification_report(y_test, predictions)}", sep="\n\n")
```
